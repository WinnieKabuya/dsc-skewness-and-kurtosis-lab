```python
# Import required libraries
import numpy as np
import matplotlib.pyplot as plt

from scipy.stats import kurtosis, skew
```

## Take 1
* Generate a random normal variable `x_random` in NumPy with 10,000 values. Set the mean value to 0 and the standard deviation to 2.
* Plot a histogram of the data, set bins to `auto` (default). 
* Calculate the skewness and kurtosis for this data distribution using the SciPy functions.
* Record your observations about the calculated values and the shape of the data. 

```python
x_random = np.random.normal(0, 2, 10000)
plt.hist(x_random, bins='auto')
print ('Skewness =', skew(x_random))
print ('kurtosis =', kurtosis(x_random))

# Skewness = -0.0025781248189666343
# kurtosis = 0.03976806960642154
```

# Your observations here 

It is typical for data to have a symmetrical distribution, which means it is normally distributed.
The data distribution is slightly skewed to the negative side, indicating that there are slightly more values on the right of the mean than on the left.
According to the kurtosis value, the data distribution is Platykurtic, which means that it has a lower-than-normal peak and wider-than-normal tails. This value was obtained using Fisher's implementation of kurtosis, which subtracts 3 from the Pearson kurtosis and is also known as excess kurtosis.
The data has a light tail and no outliers, meaning that the majority of the data is concentrated around the mean and there are no extreme values.

## Take 2

Let's generate another distribution 

```python
x = np.linspace( -5, 5, 10000 )
y = 1./(np.sqrt(2.*np.pi)) * np.exp( -.5*(x)**2  )  # normal distribution
```

* Plot a histogram for data $y$, and set bins to auto (default).
* Calculate the skewness and kurtosis for this data distribution using the SciPy functions.
* Record your observations about the calculated values and the shape of the data.

```python
plt.hist(y, bins='auto')
print ('Skewness =', skew(y))
print ('kurtosis =', kurtosis(y))



# Skewness = 1.109511549276228
# kurtosis = -0.31039027765889804
```

# Your observations here 
The distribution is highly skewed to the positive side, indicating that there are more values on the left of the mean than on the right.
The negative kurtosis value of the distribution implies that it has thinner tails and a flatter peak compared to the normal distribution. This is known as platykurtic. It's worth noting that kurtosis values are only meaningful when the observed curve is bell-shaped, and when we want to compare the tails to those of a normal distribution.
The observed distribution is clearly non-normal, as seen in the plot. The negative kurtosis value suggests that it has thinner tails and a flatter peak than the normal distribution. It's important to keep in mind that kurtosis values are most informative when the distribution is bell-shaped and we want to assess the fatness or thinness of the tails compared to a normal distribution.
